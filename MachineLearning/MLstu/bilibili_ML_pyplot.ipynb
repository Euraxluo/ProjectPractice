## matplotlib 基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
siny = np.sin(x)
cosy = np.cos(x)
```

```python
plt.plot(x,siny,label = "sin(x)")
plt.plot(x,cosy,color = "red",linestyle = ":",label = "cos(x)")
plt.axis([-1,11,-2,2])#调整x，y轴的范围
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()#必须要加上label，才能添加图示
plt.title("sin&cos")
plt.show()
```

### Scatter Plot{散点图}

```python
plt.scatter(x,siny)
plt.scatter(x,cosy)
plt.show()
```

```python
x = np.random.normal(0,1,1000)
y = np.random.normal(0,1,1000)
plt.scatter(x,y,alpha = 0.5)#二维正态分布
plt.show()
```
